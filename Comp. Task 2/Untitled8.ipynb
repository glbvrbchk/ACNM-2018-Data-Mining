```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

f = "/Users/glebvorobcuk/CLionProjects/untitled/cmake-build-debug/plot.txt"
grid = open(f,"r")
grr=np.loadtxt(grid)



f1 = "/Users/glebvorobcuk/CLionProjects/untitled/cmake-build-debug/particles.txt"
f2 = "/Users/glebvorobcuk/CLionProjects/untitled/cmake-build-debug/particles_wc2.txt"
f3 = "/Users/glebvorobcuk/CLionProjects/untitled/cmake-build-debug/particles_wc6.txt"



snap_h_c = open(f1,"r")
snap_h_wc2 = open(f2,"r")
snap_h_wc6 = open(f3,"r")




#get h
H=[]
RHO=[]
for h in snap_h_c.read().split("\n"):
    if "H:" in h:
        H.append(h[h.find("H:"):].split("H:")[1])        
#get central particle density in snapshot for each h
snap_data=pd.read_csv(f1,skiprows=2,sep="\t")

snap_df=pd.DataFrame(snap_data)


central_df=pd.DataFrame((snap_df[(snap_df['X:']=='0.515625')&(snap_df['Y:']=='0.515625')]),dtype=float)
RHO=central_df['Density:']
print("cubic")
print(H)
print(RHO)

#get h
H2=[]
RHO2=[]
for h2 in snap_h_wc2.read().split("\n"):
    if "H:" in h2:
        H2.append(h2[h2.find("H:"):].split("H:")[1])
#get central particle density in snapshot for each h
snap_data_2=pd.read_csv(f2,skiprows=2,sep="\t")

snap_df_2=pd.DataFrame(snap_data_2)

central_df_2=pd.DataFrame((snap_df_2[(snap_df_2['X:']=='0.515625')&(snap_df_2['Y:']=='0.515625')]),dtype=float)

RHO2=central_df_2['Density:']
print(H2)
print(RHO2)


#get h
H3=[]
RHO3=[]
for h3 in snap_h_wc6.read().split("\n"):
    if "H:" in h3:
        H3.append(h3[h3.find("H:"):].split("H:")[1])
#get central particle density in snapshot for each h
snap_data_3=pd.read_csv(f3,skiprows=2,sep="\t")

snap_df_3=pd.DataFrame(snap_data_3)

central_df_3=pd.DataFrame((snap_df_3[(snap_df_3['X:']=='0.515625')&(snap_df_3['Y:']=='0.515625')]),dtype=float)

RHO3=central_df_3['Density:']
print(H3)
print(RHO3)

```

```python
plt.figure(figsize=(20,20))
plt.plot(H[3:],RHO[3:],label='cubic')
plt.plot(H2[3:],RHO2[3:],label='wc2')
plt.plot(H3[3:],RHO3[3:],label='wc6')
plt.legend(['cubic', 'wc2','wc6'])
#plt.ylim(ymin=0.99,ymax=1.02)
plt.ylim(ymin=0.99,ymax=1.02)
plt.show()
```
