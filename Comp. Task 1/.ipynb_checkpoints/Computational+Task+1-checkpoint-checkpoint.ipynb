```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math as math
from scipy import stats

#import datetime as dt
#code the functions for all of the 
#'/Users/glebvorobcuk/Desktop/ACNM 2018/Term 2/Data Mining/Comp. Task 3/AMZN_.csv'

def Task_2(x):
    #SELECT THE DATA 
    AMAZON = pd.read_csv(str(x))

    AMAZON_ = AMAZON[['Date', 'Close', 'Weekday', 'OP/CL', 'Missed Weekday', 'Missed date']]


    print(AMAZON_)

    AMAZON_.Date = pd.to_datetime(AMAZON_.Date, format='%d/%m/%Y',utc=True)
    AMAZON_['Missed date']= pd.to_datetime(AMAZON_['Missed date'], format='%d/%m/%Y',utc=True)

    #make copy for normalizaton

    AMAZON_2 = pd.read_csv(str(x))
    AMAZON_2 = AMAZON_2[['Date', 'Close', 'Weekday', 'OP/CL', 'Missed Weekday', 'Missed date']]
    AMAZON_2.Date = pd.to_datetime(AMAZON_2.Date, format='%d/%m/%Y',utc=True)
    AMAZON_2['Missed date']= pd.to_datetime(AMAZON_2['Missed date'], format='%d/%m/%Y',utc=True)





    plt.figure(figsize=(20,20))
    plt.plot(AMAZON_.Date,AMAZON_.Close, color='red')

    #find missing
    #

    AMAZON_Missed = AMAZON_[AMAZON_['Missed date'].notnull()]
    AMAZON_Missed.reset_index(inplace=True)
    AMAZON_Missed['Date'] = pd.to_datetime(AMAZON_Missed['Date'] , format='%d/%m/%Y',utc=True)
    AMAZON_Missed['Missed date'] = pd.to_datetime(AMAZON_Missed['Missed date'] , format='%d/%m/%Y',utc=True)




    size = len(AMAZON_Missed.index)

    j=0
    for i in range(size):
        new_date = AMAZON_Missed.loc[i]
        data={'Date':new_date['Missed date'], 'Close':AMAZON_['Close'].loc[AMAZON_Missed['index'].loc[i]-1], 'Weekday':AMAZON_['Weekday'].loc[AMAZON_Missed['index'].loc[i]-1], 'OP/CL':AMAZON_['OP/CL'].loc[AMAZON_Missed['index'].loc[i]-1], 'Missed Weekday':0, 'Missed date':0}


        missed_line = pd.DataFrame(data=data,index=[0])

        missed_line=missed_line[['Date', 'Close', 'Weekday', 'OP/CL', 'Missed Weekday', 'Missed date']]



        missed_line =pd.DataFrame(data=data,index=[0])
        if AMAZON_Missed['OP/CL'].loc[i] == 'OP':
            AMAZON_ = pd.concat([AMAZON_.loc[:(AMAZON_Missed['index'].loc[i]-1+j)],missed_line,AMAZON_.loc[(AMAZON_Missed['index'].loc[i]+j):]]).reset_index(drop=True)
        elif AMAZON_Missed['OP/CL'].loc[i] == 'CL':
            AMAZON_ = pd.concat([AMAZON_.loc[:(AMAZON_Missed['index'].loc[i]+j)],missed_line, AMAZON_.loc[(AMAZON_Missed['index'].loc[i]+1+j):]]).reset_index(drop=True)
        j=j+1



    plt.plot(AMAZON_.Date,AMAZON_.Close,color='blue')
    plt.legend(['Before Preprocessing','After Preprocessing'])
    plt.xlabel('Date')
    plt.ylabel('Price,$')
    plt.title('Nvidia Time Series')
    plt.savefig('amazon_filled.png')

    AMAZON_2['Log-return'] = np.log(AMAZON_2.Close/AMAZON_2.Close.shift(1))
    AMAZON_2_VAR=AMAZON_2['Log-return'].var()
    AMAZON_2_MEAN=AMAZON_2['Log-return'].mean()
    #AMAZON_2_STD = AMAZON_2['Log-return'].std()



    #z-score
    zscore=pd.DataFrame(stats.zscore(AMAZON_2['Log-return'].loc[1:]))
    nan=pd.DataFrame(data={0:float('NaN')},index=[0])
    zscore = pd.concat([nan,zscore])
    zscore.reset_index(inplace=True,drop=True)
    print(zscore)
    AMAZON_2.insert(loc=7,column='Normalized',value=zscore)
    print("VARIANCE",AMAZON_2_VAR)
    print("MEAN",AMAZON_2_MEAN)
    print(AMAZON_2)
    plt.figure(figsize=(12,12))
    plt.plot(AMAZON_2.Date.loc[1:],AMAZON_2['Log-return'].loc[1:])
    plt.legend(['Log-Return AMD'])
    plt.savefig('log_ebay.png')
    plt.figure(figsize=(12,12))
    plt.plot(AMAZON_2.Date.loc[1:],AMAZON_2.Normalized.loc[1:])
    plt.legend(['Normalized AMD'])
    plt.savefig('norm_ebay.png')
    



    plt.figure(figsize=(12,12))
    ax = AMAZON_2.Normalized.plot(x='ZScore', y='FreqDist', kind='kde', figsize=(10, 6))
    arr = ax.get_children()[0]._x
    plt.xticks(np.linspace(arr[0], arr[-1]), rotation=90)
    plt.show()
    
    plt.figure()
    plt.plot(AMAZON_2.Date.loc[10:40],AMAZON_2.Normalized.loc[10:40])


```

```python

%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math as math
from scipy import stats
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

treshold=0.25
def Task_3(x,treshold):
    g = pd.read_csv(str(x))
    g=g[['Date','Close']]
    g.Date = pd.to_datetime(g.Date, format='%d/%m/%Y',utc=True)
    g['Log-return'] = np.log(g.Close) - np.log(g.Close.shift(1))
    zscore=pd.DataFrame(stats.zscore(g['Log-return'].loc[1:]))
    nan=pd.DataFrame(data={0:float('NaN')},index=[0])
    zscore = pd.concat([nan,zscore])
    zscore.reset_index(inplace=True,drop=True)
    g.insert(loc=3,column='Normalized',value=zscore)
    plt.scatter(range(0,755),g.Normalized)
    print(g)
    size = len(g.index)
    b=pd.DataFrame()
    d=pd.DataFrame()


    seg_p=pd.DataFrame()
    seg_MSE=np.array([])
    seg_to_pairs=np.array([])
    for i in range(1,size-1):
        cc=[g.Normalized.loc[i],g.Normalized.loc[i+1]]
        dd=[i,i+1]
        data={"Segments":[cc]}
        data_2={"Indices":[dd]}
        a=pd.DataFrame(data) 
        c=pd.DataFrame(data_2)
        b=b.append(a)
        d=d.append(c)
        b.reset_index(inplace=True,drop=True)
        d.reset_index(inplace=True,drop=True)

    b=b.iloc[::2]
    d=d.iloc[::2]

    b.reset_index(inplace=True,drop=True)
    d.reset_index(inplace=True,drop=True)
    b=pd.concat([b, d], axis=1)
    print("b-------")
    print(b)
    scatt = b.copy(deep=True)

    plt.figure()
    size = len(b.index)
        #size=10
    plt.figure(figsize=(40,10))
    for j in range(size):
        vvv=list(b.loc[j])[0]
        
        ddd=list(d.loc[j])
       
        plt.scatter(ddd,vvv)
        plt.plot(ddd[0],vvv)
    plt.savefig('start.png')
    plt.figure(figsize=(40,10))
    for j in range(size):
        vvv=list(b.loc[j])[0]
        
        ddd=list(d.loc[j])
       
        plt.scatter(ddd,vvv)
        plt.plot(ddd[0],vvv)
    
    
    Error=0
    iteration=0
    while Error<treshold:
        size = len(b.index)
        seg_p=pd.DataFrame()
        seg_MSE=[]
        for l in range(size-1):
            #prepare y for splitting    

            #print(b.loc[l:l+1])
            y=list(b.loc[l])[0]
            y=y+list(b.loc[l+1])[0]

            y=np.asarray(y)
            seg_data={"Pairs":[y]}
            seg=pd.DataFrame(data=seg_data)
            seg_p=seg_p.append(seg)
            seg_p.reset_index(inplace=True,drop=True)

                    #prepare x for splitting
            X=list(b.loc[l][1])
            X=X+list(b.loc[l+1])[1]
            X=np.asarray(X)



            regr = linear_model.LinearRegression()
            regr.fit(X.reshape(-1,1),y.reshape(-1,1))
            y_pred=regr.predict(X.reshape(-1,1))

            MSE=mean_squared_error(y, y_pred)
            seg_MSE=np.append(seg_MSE,MSE)
            seg_p['MSE']=seg_MSE


                    #print("Mean squared error: %.2f"% MSE)

                    # Plot outputs
                    #plt.scatter(X, y)
            plt.plot(X, y_pred, color='blue', linewidth=3)
        #plt.savefig('reg.png')


                            #bottom-up
                            #find min mse and index
        minimal=seg_p['MSE'].argmin()
                            #find pair
        Error=seg_p['MSE'].loc[minimal]
        #print("Min Index",minimal)
        #print("Min Merge Cost:",seg_p['MSE'].loc[minimal])
        #change 
        #print('Segments to merge')
        #print(b.loc[seg_p.index.get_loc(minimal)])
        #print(b.loc[seg_p.index.get_loc(minimal)+1])
        print('------')
        b.at[seg_p.index.get_loc(minimal),'Segments']=[seg_p.loc[minimal][1][0],seg_p.loc[minimal][1][3]]
        b.at[seg_p.index.get_loc(minimal),'Indices']=[b.loc[seg_p.index.get_loc(minimal)][1][0],b.loc[seg_p.index.get_loc(minimal)+1][1][1]]
        

        #print(b.loc[seg_p.index.get_loc(minimal)+1])
        b=b.drop([b.index[seg_p.index.get_loc(minimal)+1]])
        b.reset_index(inplace=True,drop=True)
        #print('After merging')
        #print(b.loc[seg_p.index.get_loc(minimal)])
        new_size=len(b.index)
        iteration=iteration+1
        print("Iteration:",iteration)
    plt.figure(figsize=(40,10))
    #print("scatt",scatt)
    for j in range(len(scatt.index)):
        vvv=list(scatt.loc[j])[0]
        ddd=list(d.loc[j])
       
        plt.scatter(ddd,vvv)
        plt.plot(ddd[0],vvv,'ro')
   

    for j in range(new_size):
        vvv=list(b.loc[j])[0]
        ddd=list(b.loc[j])[1]
        print(ddd)
        
        plt.plot(ddd,vvv)
    plt.savefig('f2.png')


    #print(seg_p)
    #print(seg_p.loc[minimal])
    #print(Error)
    print(iteration)
```

```python
x='/Users/glebvorobcuk/Desktop/ACNM 2018/Term 2/Data Mining/Comp. Task 3/EBAY_.csv'
y='/Users/glebvorobcuk/Desktop/ACNM 2018/Term 2/Data Mining/Comp. Task 3/NVDA_.csv'
z='/Users/glebvorobcuk/Desktop/ACNM 2018/Term 2/Data Mining/Comp. Task 3/AMD_.csv'
w='/Users/glebvorobcuk/Desktop/ACNM 2018/Term 2/Data Mining/Comp. Task 3/AMZN_.csv'



from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split


def Normalized(x):
    g = pd.read_csv(str(x))
    g=g[['Date','Close']]
    g.Date = pd.to_datetime(g.Date, format='%d/%m/%Y',utc=True)
    g['Log-return'] = np.log(g.Close) - np.log(g.Close.shift(1))
    zscore=pd.DataFrame(stats.zscore(g['Log-return'].loc[1:]))
    nan=pd.DataFrame(data={0:float('NaN')},index=[0])
    zscore = pd.concat([nan,zscore])
    zscore.reset_index(inplace=True,drop=True)
    g.insert(loc=1,column='Normalized',value=zscore)
    return g[['Date','Normalized']]

g = Normalized(x)

d_1=pd.DataFrame(Normalized(y))#NVDA
d_2=pd.DataFrame(Normalized(z))#AMD
d_3=pd.DataFrame(Normalized(w))#AMZN

g.insert(loc=2,column='NVDA',value=d_1.Normalized)
g.insert(loc=3,column='AMD',value=d_2.Normalized)
g.insert(loc=4,column='AMZN',value=d_1.Normalized)
g.drop([0],inplace=True)
g.insert(loc=5,column='G+1',value=g.Normalized.shift(-1))
g.reset_index(drop=True,inplace=True)


print(g.iloc[:,1:5])
print(np.isnan(g.iloc[:,1:5]))
regr = linear_model.LinearRegression()
regr.fit(g.iloc[:,1:5],g['G+1'])
y_pred=regr.predict(g['Normalized','NVDA','AMD','AMZN'])

print(g)
g.to_csv('out.csv')




```
